```python
import numpy as np

df.loc[:, 'ship-country'] = df['ship-country'].replace(['', ' ', 'None', 'nan', 'NaN'], np.nan)

cols_to_fill = ['ship-city', 'ship-state', 'ship-country', 'Category']
df.loc[:, cols_to_fill] = df[cols_to_fill].fillna('Unknown')

df.loc[:, 'ship-state'] = df['ship-state'].replace({
    'rajsthan': 'Rajasthan',
    'Rajsthan': 'Rajasthan',
    'Orissa': 'Odisha'
})

print(" Missing values after cleaning:\n")
print(df.isnull().sum())
```

```python
daily_sales = df.groupby(df['Date'].dt.date)['Amount'].sum().reset_index()

plt.figure(figsize=(12, 6))
daily_sales.plot(x='Date', y='Amount', kind='line', marker='o') 
plt.title('Daily Sales Amount')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('month')['Amount'].sum().reset_index()
plt.figure(figsize=(12, 6))
monthly_sales.plot(x='month', y='Amount', kind='bar', color='skyblue')
plt.title('Monthly Sales Amount')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
category_sales = df.groupby('Category')['Amount'].sum()
print(category_sales)
plt.figure(figsize=(12, 6))
category_sales.plot(kind='bar', color='lightgreen')
plt.title('Sales Amount by Category')
plt.xlabel('Category')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
state_sales = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=True)
plt.figure(figsize=(12, 6))
state_sales.plot(kind='bar', color='coral')
plt.title('Sales Amount by State')
plt.ylabel('State')
plt.xlabel('Sales Amount')
plt.tight_layout()
plt.show()
```

```python
top_products = df.groupby('Style')['Qty'].sum().sort_values(ascending=False).head(20)
plt.figure(figsize=(12, 6))
top_products.plot(kind='barh', color='purple')
plt.title('Top 10 Products by Quantity Sold')
plt.xlabel('Product Style')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.tight_layout()      
plt.show()
```

```python
top_categories = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
top_categories.plot(kind='bar', color='orange')
plt.title('Top 10 Categories by Sales Amount')
plt.xlabel('Category')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_states = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(20)
plt.figure(figsize=(12, 6))
top_states.plot(kind='bar', color='teal')
plt.title('Top 10 States by Sales Amount')
plt.xlabel('State')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
status_counts = df['Status'].value_counts()
plt.figure(figsize=(10, 6)) 
status_counts.plot(kind='bar', color='lightblue')
plt.title('Order Status Counts')
plt.xlabel('order Status')
plt.ylabel('number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.groupby('Fulfilment')['Amount'].sum().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'])
plt.title('Sales Amount by Fulfilment Type')
plt.ylabel('type of Fulfilment')
plt.tight_layout()
plt.show()
```
